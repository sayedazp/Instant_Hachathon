```python
def punc_remove(mylist , no_puncs = []):
    punc = string.punctuation
    if len(no_puncs) > 0:
        for pun in no_puncs:
            punc = punc.replace(pun , "")
    for char in punc:
        return [w.translate(str.maketrans('', '', punc)) for w in mylist]
    
```

# create a cleaning pipelie

```python
def split_show(data):
    x = data.split()
    print(x[0:30])
    return data.split()
```

```python
def remove_non_en(data2):     
    return [idx for idx in data2 if not re.findall(r"[^\u0000-\u05C0\u2100-\u214F]+", idx)]
```

```python
def remove_mail(data3):
    return [re.sub(r'\S*@\S*\s?', '' , w) for w in data3]    
```

```python
def remove_references(data4):
    pattern = re.compile(r"\[(\d+)\]")
    return [re.sub(pattern, '' , w) for w in data4]
```

```python
def remove_nulls(data5):
    return [w for w in data5 if not w == ""]
```

```python
def contract(data6):
    return [contractions.fix(w) for w in data6]
```

# Testing the pipeline

```python
data2 = split_show(data)
```

```python
data3 = remove_non_en(data2)
```

```python
data4 = remove_mail(data3)
```

```python
data5 = remove_references(data4)
```

```python
data6 = remove_nulls(data5)
```

```python
data7 = contract(data6)
```
